```python
# main.py
from pyaedt_library.project import Project
import pandas as pd
import numpy as np
import os
import time
import random
import math



## random variable ##

class Parameter:

	def __init__(self) :
		self.a = 1

	def _random_choice(self,X) :
		return round(np.random.choice( np.arange( X[0] , X[1]+X[2] , X[2]) ),X[3])

	def get_random_variable(self) :

		variables = []

		# ===============
		# Range setup
		# ===============

		# variables = [["core_w1", "100mm"], ["core_l1_leg", "15mm"], ["core_l1_top", "20mm"], ["core_l2", "50mm"], ["core_h1", "50mm"], ["coil_width", "5mm"], ["space_x", "5mm"], ["space_y", "5mm"], ["space_z", "1mm"],]

		N1_range = [4, 7, 1, 0]
		self.N1 = self._random_choice(N1_range)

		self.N1 = 6

		self.LV_layer = random.choice([1, 2])
		self.HV_layer = random.choice([1, 2])

		self.LV_layer = 2
		self.HV_layer = 1

		core_w1_range = [100, 250, 1, 0]
		core_l1_top_range = [15, 30, 1, 0]
		core_l1_leg_ratio_range = [1.00, 3.00, 0.01, 2]
		core_l2_range = [50, 100, 1, 0]
		core_h1_range = [50, 120, 1, 0]

		self.core_w1 = self._random_choice(core_w1_range)
		self.core_l1_top = self._random_choice(core_l1_top_range)
		self.core_l1_leg = self.core_l1_top * self._random_choice(core_l1_leg_ratio_range)
		self.core_l2 = self._random_choice(core_l2_range)
		self.core_h1 = self._random_choice(core_h1_range)

		self.core_h1 = 80
		self.core_w1 = 120
		self.core_l1_top = 25
		self.core_l1_leg = 25 * 2.5

		LV_effective_range = [0.5, 0.9, 0.01, 2]
		LV_effective = self.core_h1 * self._random_choice(LV_effective_range)
		HV_effective_range = [0.5, 0.95, 0.01, 2]
		HV_effective = self.core_h1 * self._random_choice(HV_effective_range)


		LV_effective = self.core_h1 * 40/80
		HV_effective = self.core_h1 * 75/80
		
		

		LV_coil_ratio_range = [0.75, 0.95, 0.01, 2]
		LV_coil_ratio = self._random_choice(LV_coil_ratio_range)
		if self.LV_layer == 1 :
			self.LV_coil_width = (LV_effective*LV_coil_ratio) / (self.N1+1)
			self.LV_space_z = (LV_effective*(1-LV_coil_ratio)) / (self.N1)
		else :
			self.LV_coil_width = (LV_effective*LV_coil_ratio) / (math.ceil(self.N1/2)+1)
			self.LV_space_z = (LV_effective*(1-LV_coil_ratio)) / (math.ceil(self.N1/2))

		HV_coil_ratio_range = [0.75, 0.95, 0.01, 2]
		HV_coil_ratio = self._random_choice(HV_coil_ratio_range)
		if self.HV_layer == 1 :
			self.HV_coil_width = (HV_effective*HV_coil_ratio) / (self.N1+1)
			self.HV_space_z = (HV_effective*(1-HV_coil_ratio)) / (self.N1)
		else :
			self.HV_coil_width = (HV_effective*HV_coil_ratio) / (math.ceil(self.N1/2)+1)
			self.HV_space_z = (HV_effective*(1-HV_coil_ratio)) / (math.ceil(self.N1/2))

		LV_space_range = [5, 30, 0.1 ,1]
		self.LV_space = self._random_choice(LV_space_range)
		HV_space_range = [5, 30, 0.1 ,1]
		self.HV_space = self._random_choice(HV_space_range)

		self.LV_space = 20
		self.HV_space = 5
		
		mold_thick_range = [10, 30, 0.1, 1]
		self.mold_thick = self._random_choice(mold_thick_range)

		cp_x_range = [5, 50, 0.1, 1]
		self.cp_x = self._random_choice(cp_x_range)
		cp_y_range = [5, 50, 0.1, 1]
		self.cp_y = self._random_choice(cp_y_range)
		cp_z1_range = [10, 30, 0.1, 1]
		self.cp_z1 = self._random_choice(cp_z1_range)
		cp_z2_range = [0, 0.5, 0.1, 1]
		self.cp_z2 = self._random_choice(cp_z2_range)

		length = self.LV_space + self.HV_space + self.mold_thick + self.LV_coil_width + self.HV_coil_width
		if self.core_l2 - length < 20 :
			self.get_random_variable()


time_table = []
time_table.append(time.time())

v = Parameter()
v.get_random_variable()



d = Project()
d.version = "2023.1"
d.dir = os.path.join(os.getcwd(), "test")
d.dir = "C:\\Users\\NEC_v\\Desktop\\TR_simulation"
# d.non_graphical = "False"


desktop = d.desktop()
design1 = d.new_design(design_name="test1", design_type="Maxwell3D", solver="EddyCurrent")

err = None

time_table.append(time.time())

# setup
setup = design1.create_setup(setupname = "Setup1")
setup.props["MaximumPasses"] = 5
setup.props["MinimumPasses"] = 1
setup.props["PercentError"] = 3
setup.props["Frequency"] = f'{30}kHz'



# set variables
variables = [["N1", f'{v.N1}'], ["core_w1", f'{v.core_w1}mm'], ["core_l1_top", f'{v.core_l1_top}mm'], ["core_l1_leg", f'{v.core_l1_leg}mm'], ["core_l2", f'{v.core_l2}mm'], ["core_h1", f'{v.core_h1}mm'], 
			["LV_width", f'{v.LV_coil_width}mm'], ["HV_width", f'{v.HV_coil_width}mm'], ["LV_space_z", f'{v.LV_space_z}mm'], ["HV_space_z", f'{v.HV_space_z}mm'],
			["LV_space", f'{v.LV_space}mm'], ["HV_space", f'{v.HV_space}mm'],
			["cp_x", f'{v.cp_x}mm'], ["cp_y", f'{v.cp_y}mm'], ["cp_z1", f'{v.cp_z1}mm'], ["cp_z2", f'{v.cp_z2}']]
design1.variable.set_variable(variables)

# make ferrite core
# core_a = 0.243557956
# core_x = 2.200512938
# core_y = 2.299301017
design1.excitation.set_power_ferrite(cm=0.2435*1e-3, x=2.2, y=2.299)

time_table.append(time.time())

core_variables = {"w1": "core_w1", "l1_leg": "core_l1_leg", "l1_top": "core_l1_top", "l2": "core_l2", "h1": "core_h1", "mat": "power_ferrite", "coreloss": True}
core = design1.model3D.create_coretype_core(**core_variables)

# make cold plate
cold_plate_variables = {"w1": "core_w1", "l1_leg": "core_l1_leg", "l1_top": "core_l1_top", "l2": "core_l2", "h1": "core_h1", "mat": "aluminum",
						"cold_plate_x": "cp_x", "cold_plate_y": "cp_y", "cold_plate_z1": "cp_z1", "cold_plate_z2": "cp_z2"}
cold_plate = design1.model3D.create_cold_plate(**cold_plate_variables)
[cp_u, cp_d] = cold_plate



# make region
region = design1.model3D.create_region(x_p=0, x_n=0, y_p=100, y_n=100, z_p=100, z_n=100)

# make LV winding
winding1_variables = {"core_x": "core_w1", "core_y": "core_l1_leg", "space_x": "LV_space", "space_y": "LV_space", "space_z": "LV_space_z", "coil_width": "LV_width"}
winding1 = design1.model3D.create_winding(name = "LV_winding", turns=v.N1, layer=2, **winding1_variables)
winding1.move(["0", "-(core_l1_leg + core_l2)/2", "0"])

# make HV winding
winding2_variables = {"core_x": "core_w1", "core_y": "core_l1_leg", "space_x": "HV_space", "space_y": "HV_space", "space_z": "HV_space_z", "coil_width": "HV_width"}
winding2 = design1.model3D.create_winding(name = "HV_winding", turns=v.N1, layer=1, **winding2_variables)
winding2.mirror(position = [0, 0, 0], vector = [0, -1, 0])
winding2.mirror(position = [0, 0, 0], vector = [1, 0, 0])
winding2.move(["0", "(core_l1_leg + core_l2)/2", "0"])

time_table.append(time.time())

# excitation
LV_winding = design1.excitation.assign_winding_coil(winding_obj=winding1, ter_name=["LV_in","LV_out"], winding_name="LV", rms_current=100)
HV_winding = design1.excitation.assign_winding_coil(winding_obj=winding2, ter_name=["HV_in","HV_out"], winding_name="HV", rms_current=100)
design1.assign_matrix(sources=[LV_winding.name,HV_winding.name],matrix_name="Matrix1")


# meshing
skin_meshing = design1.meshing.skin_depth_meshing(object=[winding1, winding2], freq=30e+3)

time_table.append(time.time())


# analyze
design1.analyze()

time_table.append(time.time())

parameters1 = []
parameters1.append(["Matrix1.L(LV,LV)","Lrx","uH"])
parameters1.append(["Matrix1.L(HV,HV)","Ltx","uH"])
parameters1.append(["Matrix1.L(LV,HV)","M","uH"])
parameters1.append(["Matrix1.CplCoef(HV,LV)","k",""])
parameters1.append(["Matrix1.L(LV,LV)*(Matrix1.CplCoef(HV,LV)^2)","Lmt","uH"])
parameters1.append(["Matrix1.L(HV,HV)*(Matrix1.CplCoef(HV,LV)^2)","Lmr","uH"])
parameters1.append(["Matrix1.L(LV,LV)*(1-Matrix1.CplCoef(HV,LV)^2)","Llt","uH"])
parameters1.append(["Matrix1.L(HV,HV)*(1-Matrix1.CplCoef(HV,LV)^2)","Llr","uH"])

data1 = design1.result.get_magnetic_parameter(name="data1", parameters=parameters1, mod="write")



parameters2 = []
parameters2.append([winding1, "EMLoss", "P_LV"])
parameters2.append([winding2, "EMLoss", "P_HV"])
parameters2.append([cp_u, "EMLoss", "P_cold_plate_top"])
parameters2.append([cp_d, "EMLoss", "P_cold_plate_bottom"])
parameters2.append([core, "CoreLoss", "P_core"])
parameters2.append([core, "B_mean", "B_mean_core"])

data2 = design1.result.get_calculator_parameter(name="data2", parameters=parameters2, mod="write")

conv_data1 = design1.result.get_convergence_report()

time_table.append(time.time())




# ===== EM simulation 2 =====

design1.meshing.initial_mesh()

# delete skin depth meshing
skin_meshing.delete()

V = 1000
Im = V/2/3.141592/30e+3/(float(data1["Lmt"])*1e-6)

# re excitation
LV_winding["Current"] = f'{Im} * sqrt(2)A'
HV_winding["Current"] = '0 * sqrt(2)A'

time_table.append(time.time())

# analyze
setup.props["PercentError"] = 1
design1.analyze()

time_table.append(time.time())

# second simulation data
data1b = design1.result.get_magnetic_parameter(name="data1", parameters=parameters1, mod="read")
data2b = design1.result.get_calculator_parameter(name="data2", parameters=parameters2, mod="read")

conv_data2 = design1.result.get_convergence_report()

time_table.append(time.time())




# copy icepak design
oProject = desktop.odesktop.SetActiveProject("script")
oDesign = oProject.SetActiveDesign(design1.design_name)
oDesign.CreateEMLossTarget("Icepak", "Setup1 : LastAdaptive", 
	[
		"NAME:DesignSetup",
		"Sim Type:="		, "Forced"
	])

# icepak design object
icepak = d.get_active_design()
icepak.variable.var = design1.variable.var

oDesign = oProject.SetActiveDesign(icepak.design_name)
oModule = oDesign.GetModule("BoundarySetup")
oModule.DeleteBoundaries(["EMLoss1"])

icepak.excitation.assign_EM_loss(name="Coreloss", objects=[core], design=design1, frequency=30e+3, loss_mul=2)
icepak.excitation.assign_EM_loss(name="Windingloss", objects=[winding1, winding2], design=design1, frequency=30e+3, loss_mul=1)


# icepak setup
icepak_setup = icepak.get_setup(name="Setup1")



icepak_setup.props["Flow Regime"] = "Turbulent"
icepak_setup.props["Include Gravity"] = True
icepak_setup.props["Solution Initialization - Z Velocity"] = "0.1m_per_sec"

icepak.variable.zero_temp()

obj_list = [core, winding1, winding2, cp_u, cp_d]
[i_core, i_LV_winding, i_HV_winding, i_cp_u, i_cp_d] = icepak.model3D.find_object(obj_list)

icepak.excitation.assign_icepak_source(assignment=[i_cp_u, i_cp_d], thermal_condition="Fixed Temperature", assignment_value="AmbientTemp", boundary_name="cold_plate")
icepak.excitation.assign_icepak_source(assignment=[i_LV_winding, i_HV_winding], thermal_condition="Fixed Temperature", assignment_value="AmbientTemp", boundary_name="winding")

time_table.append(time.time())

icepak.analyze()

time_table.append(time.time())

parameters3 = []
parameters3.append([core, "Temp_max", "Temp_max_core"])
parameters3.append([core, "Temp_mean", "Temp_mean_core"])
parameters3.append([winding1, "Temp_mean", "Temp_mean_LV_winding"])

data3 = icepak.result.get_calculator_parameter(name="data3", parameters=parameters3, mod="write")




# top_face = max(i_core.faces, key=lambda face: face.center[2])
# bottom_face = min(i_core.faces, key=lambda face: face.center[2]) 

import pandas as pd

pd.set_option('display.max_columns', None)

data = pd.concat([data1,data2,data1b,data2b,data3], axis=1)

data

time_table.append(time.time())

data
```
