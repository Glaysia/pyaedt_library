```python
import pandas as pd
import time
import os

def read_csv_safely(file_path):
    try:
        # 파일이 존재하는지 확인
        if not os.path.exists(file_path):
            print(f"{file_path} 파일이 존재하지 않습니다.")
            return 0
            
        # 파일이 다른 프로세스에 의해 사용중인지 확인하며 읽기 시도
        for attempt in range(3):  # 최대 3번 시도
            try:
                df = pd.read_csv(file_path)
                return len(df)
            except PermissionError:
                print(f"{file_path} 파일이 다른 프로세스에 의해 사용중입니다. 재시도 중...")
                time.sleep(2)  # 2초 대기 후 재시도
        
        print(f"{file_path} 파일 읽기 실패")
        return 0
        
    except Exception as e:
        print(f"{file_path} 파일 읽기 중 오류 발생: {str(e)}")
        return 0

while True:
    path1 = r"V:\jupyter\MFT_TAB\MFT_TAB\simulation_results.csv"
    path2 = r"U:\Ansys_NEC\MFT_TAB\simulation_results.csv"
    
    count1 = read_csv_safely(path1)
    count2 = read_csv_safely(path2)
    
    total = count1 + count2
    print(f"\n현재 시각: {time.strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"Path1 행 수: {count1}")
    print(f"Path2 행 수: {count2}") 
    print(f"총 행 수: {total}")
    
    time.sleep(600)  # 10분 대기
                                
```
